```python
# Simplified and well-commented clustering pipeline
import os
from pathlib import Path
import tempfile

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns  # nicer plots

from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA
from sklearn.metrics import silhouette_score
```

```python

# -----------------------------
# Settings & reproducibility
# -----------------------------
RND = 42
np.random.seed(RND)
```

```python

# -----------------------------
# 1) Prepare features for modeling
# -----------------------------
# Choose numeric and categorical columns explicitly.
# Keeping this explicit avoids ambiguity and makes the pipeline educational.
num_cols = ["Age", "Depense_Totale", "Frequence_Achat", "Visites_Mois", "Note_Client"]
cat_cols = ["Genre", "Ville", "Device", "Canal"]

# Drop ID before modeling (identifier must not influence clustering)
df_model = df.drop(columns=["ID"]).copy()

```

```python
# -----------------------------
# 2) Build preprocessor (scaling + one-hot)
# -----------------------------
# OneHotEncoder historically used argument `sparse`; newer sklearn uses `sparse_output`.
# Use a small try/except to build an encoder that works across versions.
ohe = None
try:
    # try new signature first (sklearn >= ~1.2)
    ohe = OneHotEncoder(handle_unknown="ignore", sparse_output=False)
except TypeError:
    # fallback to older signature
    ohe = OneHotEncoder(handle_unknown="ignore", sparse=False)

preprocessor = ColumnTransformer(transformers=[
    ("num", StandardScaler(), num_cols),
    ("cat", ohe, cat_cols)
], remainder="drop")

# Fit-transform the data -> returns a NumPy array ready for clustering
X = preprocessor.fit_transform(df_model)
print(f"\nPreprocessing done. X shape = {X.shape} (samples x features after encoding).")
```

```python

# -----------------------------
# 3) Find best K using silhouette score
# -----------------------------
# We test K from 2 to 8 (common small range for teaching / demos).
K_range = range(2, 9)  # 2..8

inertias = []
silhouettes = []

for k in K_range:
    km = KMeans(n_clusters=k, random_state=RND, n_init=10)
    labels = km.fit_predict(X)
    inertias.append(km.inertia_)
    # silhouette_score requires at least 2 clusters and more than 1 sample in each cluster
    try:
        sil = silhouette_score(X, labels)
    except Exception:
        sil = np.nan
    silhouettes.append(sil)

# -----------------------------
# 4) Plot Elbow + Silhouette side-by-side
# -----------------------------
plt.figure(figsize=(12, 4))

# Elbow (inertia)
plt.subplot(1, 2, 1)
plt.plot(list(K_range), inertias, marker="o")
plt.xticks(list(K_range))
plt.xlabel("Nombre de clusters K")
plt.ylabel("Inertia (somme des distances au carré)")
plt.title("Méthode du Coude (Elbow)")
plt.grid(True)


plt.tight_layout()
plt.show()
```

```python

# -----------------------------
# 5) Select K and fit final KMeans
# -----------------------------
# Choose K with highest silhouette (ignore NaN)
sil_array = np.array(silhouettes, dtype=float)
valid_idx = np.where(~np.isnan(sil_array))[0]
if len(valid_idx) > 0:
    best_idx = valid_idx[np.nanargmax(sil_array[valid_idx])]
    best_k = list(K_range)[best_idx]
else:
    # fallback (rare): use elbow heuristic -> take K=3
    best_k = 3

print(f"\nSelected K = {best_k} (best silhouette among tested values).")

kmeans_final = KMeans(n_clusters=best_k, random_state=RND, n_init=20)
df["Cluster"] = kmeans_final.fit_predict(X)  # attach cluster labels to original df

# Cluster distribution
print("\nCluster sizes:")
print(df["Cluster"].value_counts().sort_index())
```

```python

# -----------------------------
# 6) PCA for 2D visualization (only for plotting)
# -----------------------------
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X)
print("\nPCA explained variance ratio (2 components):", np.round(pca.explained_variance_ratio_, 3))

plt.figure(figsize=(8, 6))
sns.scatterplot(x=X_pca[:, 0], y=X_pca[:, 1], hue=df["Cluster"].astype(str), palette="viridis", s=80)
plt.title(f"Clients projected to 2D by PCA (K = {best_k})")
plt.xlabel("PC1")
plt.ylabel("PC2")
plt.legend(title="Cluster")
plt.grid(True)
plt.show()
```

```python
# -----------------------------
# 7) Business profiling: averages per cluster
# -----------------------------
print("\nCluster profiling (mean of numeric features):")
profiling = df.groupby("Cluster")[num_cols].mean().round(2)
profiling["Nombre_Clients"] = df.groupby("Cluster").size()
print(profiling.sort_index())
```
